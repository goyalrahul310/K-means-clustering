```python
#implementing kmeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

```python
X,y = make_blobs(n_samples=1000,n_features=2,centers=5)
```

```python
plt.figure(0)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
k = 5
clusters = {}
# color = ["red","green","blue","while","black"]
colors = ['green','red','blue','yellow','orange','pink']
for i in range(k):
    centre = 10.0*(2*np.random.random((X.shape[1],))-1)
    points = []
    cluster = { 
                'centre' : centre,
                 'points' :  points,
                 'colors'  : colors[i] }
    clusters[i] = cluster
    
print(clusters)
```

```python
plt.figure()
plt.scatter(X[:,0],X[:,1])
for ix in range(k):
    plt.scatter(clusters[ix]["centre"][0], clusters[ix]["centre"][1], color = "black", marker = "*", s=100)
plt.show()
```

```python
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
def assignptcluss(X,clusters):
    for ix in range(X.shape[0]):
        dist = []
        curr_x = X[ix]

        for kx in range(k):
            d = distance(curr_x, clusters[kx]['centre'])
            dist.append(d)

        current_cluster = np.argmin(dist)

     
        clusters[current_cluster]['points'].append(curr_x)
```

```python
assignptcluss(X,clusters)
for ix in range(5):
    print(clusters[ix]['centre'])
    print(clusters[ix]['points'])
```

```python
def updatecluster(clusters):
    for j in range(5):
        pts = np.array(clusters[j]['points'])
        if pts.shape[0] > 0:
            mean1 = pts.mean(axis=0)
            clusters[j]['centre'] = mean1
            clusters[j]['points'] = []
    
```

```python
updatecluster(clusters)
```
